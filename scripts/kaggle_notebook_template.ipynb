# GTNet Multi-Agent Trajectory Prediction Training

Training GTNet với 5 cải tiến:
1. **GAT (Graph Attention Networks)**: Learnable attention weights
2. **WTA Loss**: Winner-Takes-All loss cho multimodal prediction
3. **Adaptive Radius**: Velocity-based interaction radius
4. **Enhanced Metrics**: minADE, minFDE, MissRate
5. **Backward Compatibility**: Support old datasets

## Training Configuration

```python
# Configuration
DATA_DIR = "/kaggle/input/gtnet-dataset/processed"  # Thay đổi theo dataset của bạn
OUT_DIR = "/kaggle/working/models"
EPOCHS = 50
BATCH_SIZE = 32
LEARNING_RATE = 1e-3

# Model features
ENABLE_GAT = True
ENABLE_MULTIMODAL = True
ENABLE_ADAPTIVE_RADIUS = True
NUM_MODES = 3
NUM_ATTENTION_HEADS = 4
```

## Training - Baseline (không có cải tiến)

```python
# Train baseline model
!python scripts/kaggle_train_gtnet.py \
    --data-dir {DATA_DIR} \
    --out-dir {OUT_DIR}/baseline \
    --epochs {EPOCHS} \
    --batch-size {BATCH_SIZE} \
    --learning-rate {LEARNING_RATE} \
    --seed 42
```

## Training - GTNet Full (tất cả cải tiến)

```python
# Train full GTNet model với tất cả cải tiến
!python scripts/kaggle_train_gtnet.py \
    --data-dir {DATA_DIR} \
    --out-dir {OUT_DIR}/gtnet_full \
    --epochs {EPOCHS} \
    --batch-size {BATCH_SIZE} \
    --learning-rate {LEARNING_RATE} \
    --enable-gat \
    --enable-multimodal \
    --enable-adaptive-radius \
    --num-modes {NUM_MODES} \
    --num-attention-heads {NUM_ATTENTION_HEADS} \
    --seed 42
```
